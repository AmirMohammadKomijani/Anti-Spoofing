```python
import os
import glob

# Specify the folder path
folder_path = 'Dataset/Real/'

# Use glob to get all text files in the folder
text_files = glob.glob(os.path.join(folder_path, '*.txt'))

# Delete each text file
for file_path in text_files:
    try:
        os.remove(file_path)
        print(f"Deleted {file_path}")
    except Exception as e:
        print(f"Error deleting {file_path}: {e}")
```

```python
import os
import glob
import shutil

# Specify the source and destination folders
source_folder = 'data/train_img/train_img/color/'
destination_folder = 'data/real'

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Use glob to get all text files with 'fake' in their name in the source folder
text_files = glob.glob(os.path.join(source_folder, '*real*.jpg'))

# Move each text file to the destination folder
for file_path in text_files:
    try:
        shutil.move(file_path, destination_folder)
        print(f"Moved {file_path} to {destination_folder}")
    except Exception as e:
        print(f"Error moving {file_path}: {e}")
```
